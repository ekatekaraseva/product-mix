```python
import pandas as pd
import numpy as np
from  datetime import datetime
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('UKD_dostavka_days.xlsx')
```

```python
df.ops.unique()
```

```python
df['date'] = pd.to_datetime(df['year_of_arrival'].astype(str) + '-' + df['month_of_arrival'].astype(str) + '-' + df['day_of_arrival'].astype(str))
df = df.set_index('date')
```

```python
df[df.ops==130201].resample('d').sum()['cnt']
```

```python
forecast_141022 = pd.read_excel('Прогноз 5 УКД.xlsx', sheet_name = '141022').set_index('Дата')
forecast_130201 = pd.read_excel('Прогноз 5 УКД.xlsx', sheet_name = '130201').set_index('Дата')
forecast_130202 = pd.read_excel('Прогноз 5 УКД.xlsx', sheet_name = '130202').set_index('Дата')
forecast_130203 = pd.read_excel('Прогноз 5 УКД.xlsx', sheet_name = '130203').set_index('Дата')
forecast_130206 = pd.read_excel('Прогноз 5 УКД.xlsx', sheet_name = '130206').set_index('Дата')
```

# УКД 130201

```python
pd.concat([df[df.ops==130201].resample('d').sum(),forecast_130201.iloc[:,0]],axis=1).iloc[:,[5,6]].loc['2020-07-01':'2020-07-08']
```

```python
plt.figure(figsize=(16,8))
plt.plot(df[df.ops==130201].resample('d').sum()['cnt'], label='Факт ОПС 130201')
plt.plot(forecast_130201.iloc[:,0], label='Прогноз')
#plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```
